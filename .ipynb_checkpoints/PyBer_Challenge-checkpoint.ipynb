### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() to create a new DataFrame showing the sum of the fares for each date where the indices are the city type and date.
total_fare_for_each_date = pyber_data_df.groupby(['type','date']).sum()[['fare']]
total_fare_for_each_date
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
total_fare_for_each_date = total_fare_for_each_date.reset_index()
total_fare_for_each_date
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' to get the total fares for each type of city by the date. 
total_fare_for_each_date_pivot = total_fare_for_each_date.pivot(index = 'date', columns = 'type', values = 'fare')
total_fare_for_each_date_pivot
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range_fares_df = total_fare_for_each_date_pivot.loc['2019-01-01': '2019-04-29']
date_range_fares_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
date_range_fares_df.index = pd.to_datetime(date_range_fares_df.index)
date_range_fares_df.index
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
date_range_fares_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resampled_date_range_fares_df = date_range_fares_df.resample('W').sum() 
resampled_date_range_fares_df
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
```

```python
# Plot resampled DataFrame using df.plot()
ax = plt.gca()
resampled_date_range_fares_df.plot(kind = 'line', figsize = (24,8), linewidth = 2, title = 'Total Fare by City Type', xlabel = '', ylabel = 'Fare ($USD)', legend = True, ax = ax)
ax.set_ylim(0, 2500)
ax.set_xlabel ('')
ax.set_ylabel('Fare ($USD)')
ax.legend(['Rural', 'Suburban', 'Urban'], loc = 'center')
ax.set_title("Total Fare by City Type")
plt.savefig('analysis/PyBer_fare_summary.png')
plt.show()
```
